```python
#extração de x e y
x, y = df[['temperatura']].values, df[['classification']].values
print('x:\n', x)
print('y:\n', y)
```

```python
# pré-processamento
from sklearn.preprocessing import LabelEncoder
```

```python
LabelEncoder?
```

```python
#conversão de y para valores numéricos
le = LabelEncoder()
y = le.fit_transform(y.ravel())
print('y:\n', y)
```

```python
#modelo
from sklearn.linear_model import LogisticRegression
```

```python
#classificador
clf = LogisticRegression()
clf.fit(x, y)
```

```python
from numpy import *
```

```python
import numpy as np
```

```python
#gerando 100 valores de temperatura
#linearmente espaçados entre 0 e 45
#predição em novos valores de temperatura
x_test = np.linspace(start=0., stop=45., num=100).reshape(-1, 1)

#predição desses valores
y_pred =clf.predict(x_test)
```

```python
print(y_pred)
```

```python
#conversão de y_pred para os valores orgiginais
y_pred = le.inverse_transform(y_pred)
```

```python
print(y_pred)
```

```python
print(x_test)#valores escolhidos pelo algoritmo
```

```python
#output
output = {'new_temp': x_test.ravel(),
          'new_class': y_pred.ravel()}
output = pd.DataFrame(output)
```

```python
#sistema automático
def classify_temp():
  """Classifica o input do usuário."""

  ask = True #loop continuo
  while ask:
    #input de temperatura
    temp = input("Insira a temperatura (graus celsisus):")#variavel temp

    #transformar para numpy array
    temp = np.array(float(temp)).reshape(-1, 1)#primeiro converte para um float e depois convertela para um numpy array com as dimensões que nosso modelo espera#é um vetor coluna um array com uma coluna

    #realiza classificação
    class_temp = clf.predict(temp)#após a conversão podemos  aplicar a predição deste valor # é só alimentar o nosso clf.predict#utilizar modelo clf para predizer qual que é a classificação desta temperatura

    #transformação inversa para retornar a string original
    class_temp= le.inverse_transform(class_temp)#uma vez que temos nossa classificação podemos fazer a nossa transformação inversa do nosso processamento

    #classificação
    print(f"A classificação da temperatura{temp.ravel()[0]} é:", class_temp[0])

    #perguntar 
    ask = input("Nova classificação (y/n):") == 'y'#sai do programinha sempre que clicar valor diferente de y
```

```python
#rodando programa
classify_temp()
```
